# Data Setup

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
```

```python
X = pd.get_dummies(df.drop(['spending','name'],axis=1))
y = df['spending']
```

```python
poly = PolynomialFeatures(degree=2)
```

```python
scaled_X = poly.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_X, y, test_size=0.2, random_state=42)
```

# Model Training & Selection

```python
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV
```

```python
elastic = ElasticNet()
```

```python
elastic.fit(X_train,y_train)
```

```python
yred = elastic.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mae = mean_absolute_error(y_test,yred)
rmse = mean_squared_error(y_test,yred)**0.5
#Percentage of Errors wrt y_test
mae/y_test.mean(),rmse/y_test.mean() 
```

```python
param = {
        'l1_ratio':[0,0.5,0.75,1],
        'tol':[0.0001,0.001,0.00001],
        'random_state':[42],
}
```

```python
grid = GridSearchCV(estimator=elastic,param_grid=param)
```

```python
grid.fit(X_train,y_train)
```

```python
ypred_grid = grid.predict(X_test)
```

```python
mae1 = mean_absolute_error(y_test,ypred_grid)
rmse1 = mean_squared_error(y_test,ypred_grid)**0.5
mae1/y_test.mean(),rmse1/y_test.mean() 
```

```python
import joblib
```

```python
joblib.dump(grid,'Model.pkl')
```
