# Data Setup

```python
dumm = pd.get_dummies(df)
```

```python
X = dumm.drop('stroke',axis=1)
y = dumm['stroke']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

## AdaBoost Classifier

```python
import warnings
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ada = AdaBoostClassifier(random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param = {
    'n_estimators':[50,20,100],
    'learning_rate':[1.0,10,0.5],
    'algorithm':['SAMME.R','SAMME']
}
```

```python
grid_ada = GridSearchCV(estimator=ada,param_grid=param)
```

```python
grid_ada.fit(X_train,y_train)
```

```python
grid_ada.best_estimator_
```

```python
grid_ada.best_params_
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
ypred_ada = grid_ada.predict(X_test)
```

# Results

```python
print(classification_report(y_test,ypred_ada))
```

```python
print(f"Accuracy: {round(accuracy_score(y_test,ypred_ada)*100,2)}%")
```

# Model Analysis

```python
f_imp = pd.DataFrame(index=X.columns,data=grid_ada.best_estimator_.feature_importances_,columns=['Feature Importances'])
```

```python
f_imp = f_imp.sort_values("Feature Importances")
```

```python
f_imp =f_imp[f_imp['Feature Importances']>0]
```

```python
sns.barplot(x=f_imp.index,y=f_imp['Feature Importances'])
plt.title('Feature Importances')
#plt.xticks(rotation=90)
plt.show()
```
